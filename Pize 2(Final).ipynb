```python
def pdfconverter():
    pdfname_location = input("Pls input questions pdf location:  ")
    pages_number = int(input("Input the pdf page number"))
    
    # importing required modules
    import PyPDF2
    
    # creating a pdf file object
    pdfFileObj = open(pdfname_location, 'rb')
    
    # creating a pdf reader object
    pdfReader = PyPDF2.PdfReader(pdfFileObj)
    
    # printing number of pages in pdf file
#     print(len(pdfReader.pages))
    
    # creating a page object
    pageObj = pdfReader.pages[pages_number] 
    
    # extracting text from page
#     print(pageObj.extract_text())  
    
    #Stores the cleaned pdf
    Cleaned_pdf= pageObj.extract_text()
    Cleaned_Pdf_List.append(Cleaned_pdf)
    
    # closing the pdf file object
    pdfFileObj.close()
```

```python
#Defines a function for the text generation using openai
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
        )
    return response.choices[0].message["content"]
def Generate_Questions():
    # Using cleaned pdf texts to generate questions
    text = Cleaned_Pdf_List[0]
    number_of_generated_questions= input("How many questions should be generated: ")
    print("Pls do you have related topics that similar questions should be set from? If yes the type it in the space below:")
    related_topics= input()
    prompt_2 = f"""
    Your task is to perform the following actions:
    1 - Rexpress the question above using the same language style as above with multiple choice options as above.
    2 - Give the answer to the question.
    3 - Generate {number_of_generated_questions} questions similar and in the same language style to the above text and {related_topics}  
    Text: <{text}>
    """
    response = get_completion(prompt_2)
    Cleaned_Pdf_List1.append(response)
    print("\nCompletion for prompt 2:")
    print(response)
```
